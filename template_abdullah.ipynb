```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import *
import seaborn as sns

#step 1
df_2022_Ask = pd.read_csv("XAGUSD_5 Mins_Ask_2023.01.01_2023.06.30 (1).csv")
df_2022_Bid = pd.read_csv("XAGUSD_5 Mins_Bid_2023.01.01_2023.06.30 (2).csv")

df_2022_Ask.head(4)

#step 2
df_2022 = df_2022_Bid.merge(df_2022_Ask,
                            left_on='Time (UTC)', 
                            right_on='Time (UTC)', 
                            how='outer')

df_2022

# Normalize price values by dividing by 10
price_columns = ['Open_x', 'High_x', 'Low_x', 'Close_x', 'Open_y', 'High_y', 'Low_y', 'Close_y']
for col in price_columns:
    if col in df_2022.columns:
        df_2022[col] = df_2022[col] / 10

# Shift the Min & Max prices (shift minimum to 0)
# Find the global minimum across all price columns
min_price = df_2022[price_columns].min().min()
# Shift all price values so the minimum becomes 0
for col in price_columns:
    if col in df_2022.columns:
        df_2022[col] = df_2022[col] - min_price

print(f"Price values shifted. Original minimum: {min_price * 10:.4f}, New minimum: 0.0000")
print(f"New maximum: {df_2022[price_columns].max().max():.4f}")


# Separate Inputs and Outputs matrices
# Input features: All price and volume data (Open, High, Low, Close, Volume for Bid and Ask)
input_features = ['Open_x', 'High_x', 'Low_x', 'Close_x', 'Volume _x', 
                  'Open_y', 'High_y', 'Low_y', 'Close_y', 'Volume _y']

# Create inputs matrix (X) - all features except the target
X = df_2022[input_features].values

# Output target: Next period's Close price (shifted by 1 to predict future)
# For Bid Close prediction
y_bid = df_2022['Close_x'].shift(-1).values[:-1]  # Remove last NaN value
X_bid = X[:-1]  # Remove last row to match y_bid

# For Ask Close prediction
y_ask = df_2022['Close_y'].shift(-1).values[:-1]  # Remove last NaN value
X_ask = X[:-1]  # Remove last row to match y_ask

# Alternative: Predict current Close price (if you want to use all data)
# y_bid_current = df_2022['Close_x'].values
# y_ask_current = df_2022['Close_y'].values

print(f"\nInputs Matrix (X) shape: {X.shape}")
print(f"Outputs Matrix (y_bid) shape: {y_bid.shape}")
print(f"Outputs Matrix (y_ask) shape: {y_ask.shape}")
print(f"\nInput features: {input_features}")
print(f"Output target: Next period's Close price (Bid: Close_x, Ask: Close_y)")



# Create a 3D tensor
# Define lookback window (number of previous time steps to use for prediction)
lookback_window = 60  # Use last 60 time steps (5 hours of 5-minute data)

def create_sequences(data, lookback):
    """
    Create sequences from time series data for LSTM/RNN models
    Returns 3D tensor: (samples, timesteps, features)
    """
    X_seq = []
    for i in range(lookback, len(data)):
        X_seq.append(data[i-lookback:i])
    return np.array(X_seq)

# Create 3D tensor for single unified model (predicts both Bid and Ask)
# Since both use the same input features, create one 3D tensor
X_3d = create_sequences(X_bid, lookback_window)  # Same as X_ask - both use all features

# Combine both targets for multi-output prediction
y_bid_aligned = y_bid[lookback_window:]  # Align outputs with sequences
y_ask_aligned = y_ask[lookback_window:]  # Align outputs with sequences
y_combined = np.column_stack([y_bid_aligned, y_ask_aligned])  # [Bid, Ask] for each sample

print(f"\n3D Tensor Shapes for Single Unified Model:")
print(f"X_3d shape: {X_3d.shape} (samples, timesteps, features)")
print(f"y_combined shape: {y_combined.shape} (samples, 2 outputs: [Bid_Close, Ask_Close])")
print(f"\nLookback window: {lookback_window} time steps ({lookback_window * 5} minutes)")
print(f"Ready for single LSTM model with 2 outputs to predict both Bid and Ask simultaneously!")

# ============================================================================
# LSTM MODEL - Add this code at the bottom of xagusd_analysis.py
# ============================================================================

# Install tensorflow if not installed: !pip install tensorflow
# Or use: !pip install keras (for standalone keras)

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Try tensorflow.keras first, fallback to keras if needed
try:
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import LSTM, Dense, Dropout
    from tensorflow.keras.optimizers import Adam
except ImportError:
    try:
        from keras.models import Sequential
        from keras.layers import LSTM, Dense, Dropout
        from keras.optimizers import Adam
    except ImportError:
        print("Please install tensorflow: !pip install tensorflow")
        raise

# Split data into train and test sets (80-20 split)
train_size = int(len(X_3d) * 0.8)
X_train, X_test = X_3d[:train_size], X_3d[train_size:]
y_train, y_test = y_combined[:train_size], y_combined[train_size:]

print(f"\nTrain set: {X_train.shape}, Test set: {X_test.shape}")

# Create and train LSTM model (Simplified for faster training)
model_lstm = Sequential([
    LSTM(32, return_sequences=False, input_shape=(lookback_window, len(input_features))),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dense(2)  # 2 outputs: Bid and Ask
])

model_lstm.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])

print("\nLSTM Model Summary:")
model_lstm.summary()

# Train the model (Reduced epochs and increased batch size for speed)
history_lstm = model_lstm.fit(
    X_train, y_train,
    batch_size=64,  # Increased for faster training
    epochs=10,      # Reduced from 50 to 10
    validation_split=0.2,
    verbose=1
)

# Make predictions
y_pred_lstm = model_lstm.predict(X_test)

# Calculate MSE & MAE on Test dataset
mse_bid_lstm = mean_squared_error(y_test[:, 0], y_pred_lstm[:, 0])
mae_bid_lstm = mean_absolute_error(y_test[:, 0], y_pred_lstm[:, 0])
mse_ask_lstm = mean_squared_error(y_test[:, 1], y_pred_lstm[:, 1])
mae_ask_lstm = mean_absolute_error(y_test[:, 1], y_pred_lstm[:, 1])

print(f"\nLSTM Model - Test Results:")
print(f"Bid - MSE: {mse_bid_lstm:.6f}, MAE: {mae_bid_lstm:.6f}")
print(f"Ask - MSE: {mse_ask_lstm:.6f}, MAE: {mae_ask_lstm:.6f}")

# Plotting the results' graphs
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Training history
axes[0, 0].plot(history_lstm.history['loss'], label='Train Loss')
axes[0, 0].plot(history_lstm.history['val_loss'], label='Val Loss')
axes[0, 0].set_title('LSTM Model - Loss')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_ylabel('Loss')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Plot 2: Bid predictions vs actual
axes[0, 1].plot(y_test[:, 0][:500], label='Actual Bid', alpha=0.7)
axes[0, 1].plot(y_pred_lstm[:, 0][:500], label='Predicted Bid', alpha=0.7)
axes[0, 1].set_title('LSTM - Bid Price Prediction (First 500 samples)')
axes[0, 1].set_xlabel('Sample')
axes[0, 1].set_ylabel('Price')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Plot 3: Ask predictions vs actual
axes[1, 0].plot(y_test[:, 1][:500], label='Actual Ask', alpha=0.7)
axes[1, 0].plot(y_pred_lstm[:, 1][:500], label='Predicted Ask', alpha=0.7)
axes[1, 0].set_title('LSTM - Ask Price Prediction (First 500 samples)')
axes[1, 0].set_xlabel('Sample')
axes[1, 0].set_ylabel('Price')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Plot 4: Scatter plot - Predicted vs Actual (Bid)
axes[1, 1].scatter(y_test[:, 0], y_pred_lstm[:, 0], alpha=0.5, s=1)
axes[1, 1].plot([y_test[:, 0].min(), y_test[:, 0].max()], 
                [y_test[:, 0].min(), y_test[:, 0].max()], 'r--', lw=2)
axes[1, 1].set_title('LSTM - Bid: Predicted vs Actual')
axes[1, 1].set_xlabel('Actual')
axes[1, 1].set_ylabel('Predicted')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

# ============================================================================
# CNN MODEL - Add this code at the bottom of xagusd_analysis.py
# ============================================================================

# Install tensorflow if not installed: !pip install tensorflow
# Or use: !pip install keras (for standalone keras)

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Try tensorflow.keras first, fallback to keras if needed
try:
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
    from tensorflow.keras.optimizers import Adam
except ImportError:
    try:
        from keras.models import Sequential
        from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
        from keras.optimizers import Adam
    except ImportError:
        print("Please install tensorflow: !pip install tensorflow")
        raise

# Split data into train and test sets (80-20 split)
train_size = int(len(X_3d) * 0.8)
X_train, X_test = X_3d[:train_size], X_3d[train_size:]
y_train, y_test = y_combined[:train_size], y_combined[train_size:]

print(f"\nTrain set: {X_train.shape}, Test set: {X_test.shape}")

# Create and train CNN model (Simplified for faster training)
model_cnn = Sequential([
    Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(lookback_window, len(input_features))),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=16, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(2)  # 2 outputs: Bid and Ask
])

model_cnn.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])

print("\nCNN Model Summary:")
model_cnn.summary()

# Train the model (Reduced epochs and increased batch size for speed)
history_cnn = model_cnn.fit(
    X_train, y_train,
    batch_size=64,  # Increased for faster training
    epochs=10,      # Reduced from 50 to 10
    validation_split=0.2,
    verbose=1
)

# Make predictions
y_pred_cnn = model_cnn.predict(X_test)

# Calculate MSE & MAE on Test dataset
mse_bid_cnn = mean_squared_error(y_test[:, 0], y_pred_cnn[:, 0])
mae_bid_cnn = mean_absolute_error(y_test[:, 0], y_pred_cnn[:, 0])
mse_ask_cnn = mean_squared_error(y_test[:, 1], y_pred_cnn[:, 1])
mae_ask_cnn = mean_absolute_error(y_test[:, 1], y_pred_cnn[:, 1])

print(f"\nCNN Model - Test Results:")
print(f"Bid - MSE: {mse_bid_cnn:.6f}, MAE: {mae_bid_cnn:.6f}")
print(f"Ask - MSE: {mse_ask_cnn:.6f}, MAE: {mae_ask_cnn:.6f}")

# Plotting the results' graphs
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Training history
axes[0, 0].plot(history_cnn.history['loss'], label='Train Loss')
axes[0, 0].plot(history_cnn.history['val_loss'], label='Val Loss')
axes[0, 0].set_title('CNN Model - Loss')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_ylabel('Loss')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Plot 2: Bid predictions vs actual
axes[0, 1].plot(y_test[:, 0][:500], label='Actual Bid', alpha=0.7)
axes[0, 1].plot(y_pred_cnn[:, 0][:500], label='Predicted Bid', alpha=0.7)
axes[0, 1].set_title('CNN - Bid Price Prediction (First 500 samples)')
axes[0, 1].set_xlabel('Sample')
axes[0, 1].set_ylabel('Price')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Plot 3: Ask predictions vs actual
axes[1, 0].plot(y_test[:, 1][:500], label='Actual Ask', alpha=0.7)
axes[1, 0].plot(y_pred_cnn[:, 1][:500], label='Predicted Ask', alpha=0.7)
axes[1, 0].set_title('CNN - Ask Price Prediction (First 500 samples)')
axes[1, 0].set_xlabel('Sample')
axes[1, 0].set_ylabel('Price')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Plot 4: Scatter plot - Predicted vs Actual (Bid)
axes[1, 1].scatter(y_test[:, 0], y_pred_cnn[:, 0], alpha=0.5, s=1)
axes[1, 1].plot([y_test[:, 0].min(), y_test[:, 0].max()], 
                [y_test[:, 0].min(), y_test[:, 0].max()], 'r--', lw=2)
axes[1, 1].set_title('CNN - Bid: Predicted vs Actual')
axes[1, 1].set_xlabel('Actual')
axes[1, 1].set_ylabel('Predicted')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

# ============================================================================
# MLP MODEL - Add this code at the bottom of xagusd_analysis.py
# ============================================================================

# Install tensorflow if not installed: !pip install tensorflow
# Or use: !pip install keras (for standalone keras)

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Try tensorflow.keras first, fallback to keras if needed
try:
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense, Dropout
    from tensorflow.keras.optimizers import Adam
except ImportError:
    try:
        from keras.models import Sequential
        from keras.layers import Dense, Dropout
        from keras.optimizers import Adam
    except ImportError:
        print("Please install tensorflow: !pip install tensorflow")
        raise

# Split data into train and test sets (80-20 split)
train_size = int(len(X_3d) * 0.8)
X_train, X_test = X_3d[:train_size], X_3d[train_size:]
y_train, y_test = y_combined[:train_size], y_combined[train_size:]

# Reshape 3D tensor to 2D for MLP (flatten timesteps and features)
X_train_2d = X_train.reshape(X_train.shape[0], -1)  # (samples, timesteps * features)
X_test_2d = X_test.reshape(X_test.shape[0], -1)    # (samples, timesteps * features)

print(f"\nTrain set: {X_train_2d.shape}, Test set: {X_test_2d.shape}")

# Create and train MLP model (Simplified for faster training)
model_mlp = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_2d.shape[1],)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(2)  # 2 outputs: Bid and Ask
])

model_mlp.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])

print("\nMLP Model Summary:")
model_mlp.summary()

# Train the model (Reduced epochs and increased batch size for speed)
history_mlp = model_mlp.fit(
    X_train_2d, y_train,
    batch_size=64,  # Increased for faster training
    epochs=10,      # Reduced from 50 to 10
    validation_split=0.2,
    verbose=1
)

# Make predictions
y_pred_mlp = model_mlp.predict(X_test_2d)

# Calculate MSE & MAE on Test dataset
mse_bid_mlp = mean_squared_error(y_test[:, 0], y_pred_mlp[:, 0])
mae_bid_mlp = mean_absolute_error(y_test[:, 0], y_pred_mlp[:, 0])
mse_ask_mlp = mean_squared_error(y_test[:, 1], y_pred_mlp[:, 1])
mae_ask_mlp = mean_absolute_error(y_test[:, 1], y_pred_mlp[:, 1])

print(f"\nMLP Model - Test Results:")
print(f"Bid - MSE: {mse_bid_mlp:.6f}, MAE: {mae_bid_mlp:.6f}")
print(f"Ask - MSE: {mse_ask_mlp:.6f}, MAE: {mae_ask_mlp:.6f}")

# Plotting the results' graphs
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Training history
axes[0, 0].plot(history_mlp.history['loss'], label='Train Loss')
axes[0, 0].plot(history_mlp.history['val_loss'], label='Val Loss')
axes[0, 0].set_title('MLP Model - Loss')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_ylabel('Loss')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Plot 2: Bid predictions vs actual
axes[0, 1].plot(y_test[:, 0][:500], label='Actual Bid', alpha=0.7)
axes[0, 1].plot(y_pred_mlp[:, 0][:500], label='Predicted Bid', alpha=0.7)
axes[0, 1].set_title('MLP - Bid Price Prediction (First 500 samples)')
axes[0, 1].set_xlabel('Sample')
axes[0, 1].set_ylabel('Price')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Plot 3: Ask predictions vs actual
axes[1, 0].plot(y_test[:, 1][:500], label='Actual Ask', alpha=0.7)
axes[1, 0].plot(y_pred_mlp[:, 1][:500], label='Predicted Ask', alpha=0.7)
axes[1, 0].set_title('MLP - Ask Price Prediction (First 500 samples)')
axes[1, 0].set_xlabel('Sample')
axes[1, 0].set_ylabel('Price')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Plot 4: Scatter plot - Predicted vs Actual (Bid)
axes[1, 1].scatter(y_test[:, 0], y_pred_mlp[:, 0], alpha=0.5, s=1)
axes[1, 1].plot([y_test[:, 0].min(), y_test[:, 0].max()], 
                [y_test[:, 0].min(), y_test[:, 0].max()], 'r--', lw=2)
axes[1, 1].set_title('MLP - Bid: Predicted vs Actual')
axes[1, 1].set_xlabel('Actual')
axes[1, 1].set_ylabel('Predicted')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

```

```python
"""
Ultra-Simplified Silver Forecasting Pipeline (5-minute timeframe)
----------------------------------------------------------------
- VERY lightweight LSTM / CNN / MLP so it runs on slow PCs.
- All your original hyperparameters preserved.
- Added many extra graphs for diagnostics.
- Change forecast horizon using FORECAST_STEPS.
"""

# ======================
# 1) USER PARAMETERS
# ======================
FORECAST_STEPS = 25
LOOKBACK_WINDOW = 60
MODEL_TYPE = 'CNN'      # 'LSTM' | 'CNN' | 'MLP'
EPOCHS = 5
BATCH_SIZE = 64
LEARNING_RATE = 0.001
TEST_SPLIT = 0.2
VALIDATION_SPLIT_DURING_TRAIN = 0.2
RANDOM_SEED = 42

# ======================
# 2) IMPORTS
# ======================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

np.random.seed(RANDOM_SEED)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    LSTM, Dense, Dropout, Conv1D, MaxPooling1D, Flatten, InputLayer
)
from tensorflow.keras.optimizers import Adam

# ======================
# 3) LOAD DATA
# ======================
ask_path = "XAGUSD_5 Mins_Ask_2023.01.01_2023.06.30 (1).csv"
bid_path = "XAGUSD_5 Mins_Bid_2023.01.01_2023.06.30 (2).csv"

df_ask = pd.read_csv(ask_path)
df_bid = pd.read_csv(bid_path)

time_col = "Time (UTC)"
df_ask[time_col] = pd.to_datetime(df_ask[time_col])
df_bid[time_col] = pd.to_datetime(df_bid[time_col])

df = df_bid.merge(df_ask, on=time_col, suffixes=("_bid", "_ask"))
df.sort_values(time_col, inplace=True)
df.reset_index(drop=True, inplace=True)

# ======================
# 4) PRICE + VOLUME PLOTS (4 GRAPHS)
# ======================
plt.figure(figsize=(10,4))
plt.plot(df["Close_bid"], label="Bid Close")
plt.plot(df["Close_ask"], label="Ask Close")
plt.title("Close Prices")
plt.grid(); plt.legend(); plt.show()

plt.figure(figsize=(10,4))
plt.plot(df["Open_bid"], label="Bid Open")
plt.plot(df["Open_ask"], label="Ask Open")
plt.title("Open Prices")
plt.grid(); plt.legend(); plt.show()

plt.figure(figsize=(10,4))
plt.plot(df["High_bid"], label="Bid High")
plt.plot(df["High_ask"], label="Ask High")
plt.title("High Prices")
plt.grid(); plt.legend(); plt.show()



# ======================
# 5) NORMALISATION
# ======================
price_cols = [c for c in df.columns if any(k in c.lower() for k in ["open","high","low","close"])]
vol_cols = [c for c in df.columns if "volume" in c.lower()]

df[price_cols] = df[price_cols] / 10.0
global_min = df[price_cols].min().min()
df[price_cols] = df[price_cols] - global_min

# ======================
# 6) TARGET SHIFT
# ======================
df["target_bid"] = df["Close_bid"].shift(-FORECAST_STEPS)
df["target_ask"] = df["Close_ask"].shift(-FORECAST_STEPS)
df.dropna(inplace=True)

# ======================
# 7) PREPARE MATRICES
# ======================
X_raw = df[price_cols + vol_cols].values.astype(np.float32)
y_raw = df[["target_bid", "target_ask"]].values.astype(np.float32)

def create_sequences(X, y, lookback):
    Xs, ys = [], []
    for i in range(lookback, len(X)):
        Xs.append(X[i-lookback:i])
        ys.append(y[i])
    return np.array(Xs), np.array(ys)

X, y = create_sequences(X_raw, y_raw, LOOKBACK_WINDOW)

split_idx = int(len(X) * (1 - TEST_SPLIT))
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

n_features = X.shape[2]

# ======================
# 8) SIMPLE MODELS
# ======================

def build_lstm(inp):
    model = Sequential([
        InputLayer(inp),
        LSTM(16),                 # reduced from 32 → 16
        Dropout(0.2),
        Dense(16, activation='relu'),
        Dense(2)
    ])
    return model

def build_cnn(inp):
    model = Sequential([
        InputLayer(inp),
        Conv1D(16, 3, activation='relu'),
        MaxPooling1D(2),
        Conv1D(8, 3, activation='relu'),
        Flatten(),
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(2)
    ])
    return model

def build_mlp(inp):
    flat = inp[0] * inp[1]
    model = Sequential([
        InputLayer((flat,)),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(2)
    ])
    return model

MODEL_TYPE = MODEL_TYPE.upper()

if MODEL_TYPE == "LSTM":
    model = build_lstm((LOOKBACK_WINDOW, n_features))

elif MODEL_TYPE == "CNN":
    model = build_cnn((LOOKBACK_WINDOW, n_features))

elif MODEL_TYPE == "MLP":
    X_train = X_train.reshape(X_train.shape[0], -1)
    X_test  = X_test.reshape(X_test.shape[0], -1)
    model = build_mlp((LOOKBACK_WINDOW, n_features))

optimizer = Adam(learning_rate=LEARNING_RATE)
model.compile(optimizer=optimizer, loss='mse', metrics=['mae', 'mse'])

# ======================
# 9) TRAIN MODEL
# ======================
history = model.fit(
    X_train, y_train,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=VALIDATION_SPLIT_DURING_TRAIN,
    shuffle=True
)

# ======================
# 10) PREDICT + METRICS
# ======================
y_pred = model.predict(X_test)

mse_bid = mean_squared_error(y_test[:,0], y_pred[:,0])
mae_bid = mean_absolute_error(y_test[:,0], y_pred[:,0])
mse_ask = mean_squared_error(y_test[:,1], y_pred[:,1])
mae_ask = mean_absolute_error(y_test[:,1], y_pred[:,1])

print("MSE Bid:", mse_bid)
print("MAE Bid:", mae_bid)
print("MSE Ask:", mse_ask)
print("MAE Ask:", mae_ask)

# ======================
# 11) TRAINING GRAPHS
# ======================
plt.figure(figsize=(9,4))
plt.plot(history.history['loss'], label="Train Loss")
plt.plot(history.history['val_loss'], label="Val Loss")
plt.title("Training Loss")
plt.legend(); plt.grid(); plt.show()

plt.figure(figsize=(9,4))
plt.plot(history.history['mae'], label="Train MAE")
plt.plot(history.history['val_mae'], label="Val MAE")
plt.title("Training MAE")
plt.legend(); plt.grid(); plt.show()

plt.figure(figsize=(9,4))
plt.plot(history.history['mse'], label="Train MSE")
plt.plot(history.history['val_mse'], label="Val MSE")
plt.title("Training MSE")
plt.legend(); plt.grid(); plt.show()

# ======================
# 12) MSE & MAE CURVES
# ======================
mse_curve = (y_test[:,0] - y_pred[:,0])**2
mae_curve = np.abs(y_test[:,0] - y_pred[:,0])

plt.figure(figsize=(10,4))
plt.plot(mse_curve)
plt.title("MSE Curve")
plt.grid(); plt.show()

plt.figure(figsize=(10,4))
plt.plot(mae_curve)
plt.title("MAE Curve")
plt.grid(); plt.show()

# ======================
# 13) FINAL 4 DIAGNOSTIC PLOTS
# ======================

# 1) Actual vs Pred (Bid)
plt.figure(figsize=(10,4))
plt.plot(y_test[:400,0], label="Actual")
plt.plot(y_pred[:400,0], label="Predicted")
plt.title("Actual vs Predicted (Bid)")
plt.legend(); plt.grid(); plt.show()

# 2) Scatter
plt.figure(figsize=(6,6))
plt.scatter(y_test[:,0], y_pred[:,0], s=3)
plt.title("Scatter: Pred vs Actual (Bid)")
plt.grid(); plt.show()

# 3) Error distribution
errors = y_test[:,0] - y_pred[:,0]
plt.figure(figsize=(7,4))
plt.hist(errors, bins=60)
plt.title("Error Distribution")
plt.grid(); plt.show()

# 4) Rolling 200-sample MAE
rolling = pd.Series(mae_curve).rolling(200).mean()
plt.figure(figsize=(10,4))
plt.plot(rolling)
plt.title("Rolling MAE (200 samples)")
plt.grid(); plt.show()

# ======================
# 14) SAVE WEIGHTS
# ======================
model.save_weights(f"silver_{MODEL_TYPE.lower()}.weights.h5")
print("Saved model weights.")

```
