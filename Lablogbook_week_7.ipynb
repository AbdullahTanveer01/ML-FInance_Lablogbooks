```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

df = pd.read_csv('GOLD_2022_normalised_NEW-1000.csv')
cols = ['Open_Bid', 'High_Bid', 'Low_Bid', 'Close_Bid', 'Volume_Bid', 'Open_Ask', 'High_Ask', 'Low_Ask', 'Close_Ask', 'Volume_Ask']
X = StandardScaler().fit_transform(df[cols].values)
y = df['Close_Bid'].values

def seq(X_data, y_data, n):
    return np.array([X_data[i:i+n] for i in range(len(X_data)-n)]), np.array([y_data[i+n] for i in range(len(X_data)-n)])

X_seq, y_seq = seq(X, y, 10)
X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42)

print("ORIGINAL LSTM MODEL")
print("="*60)
m1 = keras.Sequential([layers.LSTM(100, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])), layers.LSTM(100), layers.Dense(50), layers.Dense(1)])
m1.compile(optimizer='adam', loss='mse', metrics=['mae'])
m1.summary()
h1 = m1.fit(X_train, y_train, batch_size=32, epochs=5, validation_split=0.2, callbacks=[keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)], verbose=1)
mse1, mae1 = m1.evaluate(X_test, y_test, verbose=0)
print(f"Original LSTM - Test MSE: {mse1:.6f}, Test MAE: {mae1:.6f}")

print("\n\nMODIFIED LSTM MODEL")
print("="*60)
m2 = keras.Sequential([layers.LSTM(33, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])), layers.LSTM(33), layers.Dense(50), layers.Dense(1)])
m2.compile(optimizer='adam', loss='mse', metrics=['mae'])
m2.summary()
h2 = m2.fit(X_train, y_train, batch_size=32, epochs=5, validation_split=0.2, callbacks=[keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)], verbose=1)
mse2, mae2 = m2.evaluate(X_test, y_test, verbose=0)
print(f"Modified LSTM - Test MSE: {mse2:.6f}, Test MAE: {mae2:.6f}")

print("\n\nCOMPARISON")
print("="*60)
print(f"Original LSTM - MSE: {mse1:.6f}, MAE: {mae1:.6f}")
print(f"Modified LSTM - MSE: {mse2:.6f}, MAE: {mae2:.6f}")
print(f"MSE Difference: {abs(mse2-mse1):.6f}")
print(f"MAE Difference: {abs(mae2-mae1):.6f}")

plt.figure(figsize=(12, 6))
plt.plot(h1.history['mae'], label='Original Train MAE', linewidth=2, marker='o')
plt.plot(h1.history['val_mae'], label='Original Val MAE', linewidth=2, marker='s')
plt.plot(h2.history['mae'], label='Modified Train MAE', linewidth=2, marker='o')
plt.plot(h2.history['val_mae'], label='Modified Val MAE', linewidth=2, marker='s')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('MAE', fontsize=12)
plt.title('MAE Detailed Graph - LSTM Models', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
